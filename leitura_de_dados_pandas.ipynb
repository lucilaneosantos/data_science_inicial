# **Aplicação de Leitura de dados conforme extensão.**


```
import pandas as pd # importando blibioteca para leitura dos dados
```

*Leitura de dados Json* 

```
json = open('aluguel.json') # abrindo dados e realizado exibição
print(json.read())
```

```
df_json = pd.read_json('aluguel.json') # Atribuido a variavel a leitura dos dados e executando em dataFrame
df_json.head()
```

*Leitura de Texto*

```
txt = open('aluguel.txt') # abrindo dados e realizado exibição
print(txt.read())
```

```
df_txt = pd.read_table('aluguel.txt') # Atribuido a variavel a leitura dos dados e executando em dataFrame
df_txt.head()
```

*Leitura em Excel*

```
df_xlsx = pd.read_excel('aluguel.xlsx')# Atribuido a variavel a leitura dos dados e executando em dataFrame
df_xlsx
```

*Leitura de HTML*

```
df_html = pd.read_html('dados_html_1.html')# Atribuido a variavel a leitura dos dados e executando em dataFrame
df_html[0]
```

Leitura CSV

```
df_csv = pd.read_csv('aluguel.csv',sep=';')# Atribuido a variavel a leitura dos dados e executando em dataFrame
df_csv
```
